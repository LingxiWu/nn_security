```python
from pandas import * # file io
import math
from matplotlib import pyplot as plt

############################# 
######### 0u ##############
data = read_csv("1R_FS_0u_FFT.csv")

freq = data["FFT_analysis X"]
real = data["FFT_analysis YRe"]
img = data["FFT_analysis YReImag"]

# mag
mag = []
for i in range(0, len(freq)):
    mag.append(math.sqrt(real[i]**2 + img[i]**2))

# a list of tuples ([freq_0, mag_0], [freq_1, mag_1], ...,)
tup_list = []
for i in range(0, len(freq)):
    tup = (freq[i], mag[i])
    tup_list.append(tup)

# print(tup_list)
tup_list.sort(key = lambda x: x[1], reverse = True)
print(tup_list[:30])
```
