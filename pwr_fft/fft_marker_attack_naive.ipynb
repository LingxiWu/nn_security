```python
from pandas import * # file io
import math
from matplotlib import pyplot as plt
import csv

THRESHOLD = 30 # only looking at the first #_THRESHOLD frequencies
TOP = 3 # consider the top TOP FFT lib traces. 

###########
# victim FFT 
victim_fft = "7R_sweep_0p5/7R_FS_1u_4u_8u_FFT.csv"
# victim_fft = "7R_sweep_0p5/7R_FS_1_8p5_3_6_2p5_7_4_COMBO_FFT.csv"

victim_tup_list = []
with open(victim_fft, "r") as fft_file:
    csv_reader = csv.reader(fft_file, delimiter=',')
    next(csv_reader) # skip header row
    for lines in csv_reader: 
        tup = (lines[0],math.sqrt(float(lines[1])**2 + float(lines[2])**2))
        victim_tup_list.append(tup)

victim_tup_list.sort(key = lambda x: x[1], reverse = True)
# print(victim_tup_list[:THRESHOLD])


#############
# library FFT
# dir = "7R"
# files = ["7R_FS_0u_FFT.csv", "7R_FS_1u_FFT.csv", "7R_FS_2u_FFT.csv", 
#          "7R_FS_4u_FFT.csv", "7R_FS_6u_FFT.csv", "7R_FS_8u_FFT.csv"]

dir = "7R_sweep_0p5/"
files = ['7R_FS_0u_FFT.csv', '7R_FS_0p5u_FFT.csv',
         '7R_FS_1u_FFT.csv', '7R_FS_1p5u_FFT.csv', 
         '7R_FS_2u_FFT.csv', '7R_FS_2p5u_FFT.csv',  
         '7R_FS_3u_FFT.csv', '7R_FS_3p5u_FFT.csv', 
         '7R_FS_4u_FFT.csv', '7R_FS_4p5u_FFT.csv', 
         '7R_FS_5u_FFT.csv', '7R_FS_5p5u_FFT.csv', 
         '7R_FS_6u_FFT.csv', '7R_FS_6p5u_FFT.csv', 
         '7R_FS_7u_FFT.csv', '7R_FS_7p5u_FFT.csv', 
         '7R_FS_8u_FFT.csv', '7R_FS_8p5u_FFT.csv']

# read FFT library traces
lib_freq_mag = []
for i in range(len(files)):
    tup_lst = []
    with open(dir+files[i], "r") as fft_file:
        csv_reader = csv.reader(fft_file, delimiter=',')
        next(csv_reader) # skip header row
        for lines in csv_reader: 
            tup = (lines[0],math.sqrt(float(lines[1])**2 + float(lines[2])**2))
            tup_lst.append(tup)
    lib_freq_mag.append(tup_lst) # [[(freq0, mag_0), (freq1, mag1),...],[]]

# sort by magnitude    
for i in range(len(files)):
    lib_freq_mag[i].sort(key = lambda x: x[1], reverse = True)
    
# count the occurrences of victim's top frequeicies in each FFT library trace
freq_lib_list = [] # [[0, 1, 2],[2, 3, 5],[file_i, file_j, ...],...]. #_THRESHOLD entries in this list
for i in range(0, THRESHOLD):
    frequency = victim_tup_list[i][0]
    lib_lst = []
    for j in range(0, len(lib_freq_mag)): # iterate through all FFT lib traces
        for k in range(0, THRESHOLD): # for each FFT trace, iterate through top THRESHOLD frequencies
            if frequency == lib_freq_mag[j][k][0]:
                lib_lst.append(j)
    freq_lib_list.append(lib_lst)   

# print(freq_lib_list)

# freq_lib_list stores for each top frequency of the victim FFT, which lib FFT it appears: for e.g.,
# [[fft_0, fft_1, fft_12,...],[fft_2, fft_3, fft_6,...],[],...,[]]

hist = [] # [[fft_0, 5],[fft_1, 6],...,[]] for each lib FFT trace, how many times the marker appear
for i in range(0, len(files)):
    hist.append([i])


for i in range(0, len(files)): # i: lib FFT file index
    count = 0
    for j in range(0, THRESHOLD): # iterate through victim's top list
        for k in range(0, len(freq_lib_list[j])):
            if i == freq_lib_list[j][k]:
                count += 1
    hist[i].append(count)    

# print(hist)  

# extract top FFT traces from the hist
for i in range(len(files)):
    hist.sort(key = lambda x: x[1], reverse = True)

print(hist)

# 1u --> 2 
# 8p5u --> 17 
# 3u --> 6 
# 6u --> 12 
# 2p5u --> 5 
# 7u --> 14
# 4u --> 8 

# 0 -> 7R_FS_0u_FFT.csv 
# 1 -> 7R_FS_0p5u_FFT.csv 
# 2 -> 7R_FS_1u_FFT.csv 
# 3 -> 7R_FS_1p5u_FFT.csv 
# 4 -> 7R_FS_2u_FFT.csv 
# 5 -> 7R_FS_2p5u_FFT.csv  
# 6 -> 7R_FS_3u_FFT.csv 
# 7 -> 7R_FS_3p5u_FFT.csv 
    # 8 -> 7R_FS_4u_FFT.csv
# 9 -> 7R_FS_4p5u_FFT.csv 
# 10 -> 7R_FS_5u_FFT.csv 
# 11 -> 7R_FS_5p5u_FFT.csv 
# 12 -> 7R_FS_6u_FFT.csv 
# 13 -> 7R_FS_6p5u_FFT.csv 
# 14 -> 7R_FS_7u_FFT.csv
# 15 -> 7R_FS_7p5u_FFT.csv  
    # 16 -> 7R_FS_8u_FFT.csv 
    # 17 -> 7R_FS_8p5u_FFT.csv
    
```
